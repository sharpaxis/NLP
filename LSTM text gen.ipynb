```python
def read_file(filepath):
    with open(filepath) as f:
        text = f.read()
    return text
```

```python
import spacy
nlp = spacy.load('en_core_web_lg',disable=['parser','tagger','ner'])
nlp.max_length = 1198623
```

```python
def punct(text):
    doc = nlp(text)
    return [token.text.lower() for token in doc if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
md = read_file('moby_dick_four_chapters.txt')
tokens = punct(md)
```

```python
#create sequence 
train_len = 25+1
text_sequence = []
for i in range(train_len,len(tokens)):
    seq = tokens[i-train_len:i]
    text_sequence.append(seq)
```

```python
#tokenize the sequence 
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequence)
sequences = tokenizer.texts_to_sequences(text_sequence)
```

```python
vocabulary_size = len(tokenizer.word_counts)
```

```python
import numpy as np
sequences = np.array(sequences)
X = sequences[:,:-1]
y = sequences[:,-1]
seq_len = X.shape[1]
```

```python
from keras.utils import to_categorical
y = to_categorical(y)
```

```python
from tensorflow.keras.layers import Dense,LSTM,Embedding
from tensorflow.keras.models import Sequential
```

```python
def create_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size,seq_len))
    model.add(LSTM(1000, return_sequences=True))
    model.add(LSTM(1000))
    model.add(Dense(1000, activation='relu'))
    model.add(Dense(500, activation='relu'))
    model.add(Dense(250, activation='relu'))
    model.add(Dense(125, activation='relu'))
    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model
```

```python
model_2 = create_model(vocabulary_size+1,seq_len)
```

```python
model_2.fit(X,y,verbose=1,epochs=500,batch_size=256)
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
model_2.save('LSTM_mobey_dick_500.keras')
```

```python
def generate_text(model,tokenizer,seq_len,seed_text,num_gen_words):
    
    output_text = []
    
    input_text = seed_text
    
    for i in range(num_gen_words):
        #encode the input text to pass into model
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        #as we trained model on 25 tokens, pad the seed text to make sure it matches lenghth
        pad_encoded = pad_sequences([encoded_text],maxlen=seq_len,truncating='pre')
        #predicting the word
        prediction = model.predict(pad_encoded,verbose=0)
        #index of thw word
        pred_word_ind = np.argmax(prediction,axis=1)[0]
        pred_word = tokenizer.index_word[pred_word_ind]
        #add the predicted word to input
        input_text += ' '+pred_word
        output_text.append(pred_word)
    return ' '.join(output_text)
```
